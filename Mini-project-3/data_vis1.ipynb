```python
import pandas as pd
```

```python
df = pd.read_csv("./data.csv")
```

```python
df
```

```python
data_r = df[["recipient","donor","commitment_amount_usd_constant"]]
```

```python
data_r1 = data_r.groupby(['recipient'], as_index=False).agg({"commitment_amount_usd_constant": "sum"})
```

```python
data_agg = data_r1.sort_values(by=['commitment_amount_usd_constant'],ascending=False)
```

```python
top_10r = data_agg.nlargest(10, 'commitment_amount_usd_constant')
```

```python
top_10r
```

```python
a = list(top_10r['recipient'])
```

```python
a1 = a.reverse()
```

```python
a1
```

```python
data_r2 = data_r.groupby(['donor'], as_index=False).agg({"commitment_amount_usd_constant": "sum"})
```

```python
data_agg2 = data_r2.sort_values(by=['commitment_amount_usd_constant'],ascending=False)
```

```python
top_20d = data_agg2.nlargest(20, 'commitment_amount_usd_constant')
```

```python
b = list(top_20d['donor'])
```

```python
b
```

```python
data = data_r[data_r.recipient.isin(a)]
```

```python
data1 = data[data.donor.isin(b)]
```
