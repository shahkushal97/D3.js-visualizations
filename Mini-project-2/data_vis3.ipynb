```python
import pandas as pd
```

```python
df = pd.read_csv("./aiddata.csv")
```

```python
df = df.drop(['aiddata_id','aiddata_2_id'],axis=1)
```

```python
def process_data(df):
    dict_donation = {}
    dict_receiver = {}
    for index, row in df.iterrows():
        try:
            dict_donation[row["donor"]] += row["commitment_amount_usd_constant"]
        except:
            dict_donation[row["donor"]] = row["commitment_amount_usd_constant"]

        try:
            dict_receiver[row["recipient"]] += row["commitment_amount_usd_constant"]
        except:
            dict_receiver[row["recipient"]] = row["commitment_amount_usd_constant"]

    donated = set(dict_donation.keys())
    received = set(dict_receiver.keys())

    not_donated = received - donated

    result = list(dict_donation.keys()) + list(not_donated) 

    df_aid = pd.DataFrame(columns=['country', 'donated', 'received'])
    for i in range(len(result)):

        country = result[i]
        try:
            donated = dict_donation[country]
        except:
            donated = 0.0

        try:
            received = dict_receiver[country]
        except:
            received = 0.0
        df_aid.loc[i] = [country, donated, received]
    return df_aid
```

```python
file1_df = process_data(df[df['year'] <= 1980])
file2_df = process_data(df[(df['year'] > 1980) & (df['year'] <= 1990)])
file3_df = process_data(df[(df['year'] > 1990) & (df['year'] <= 2000)])
file4_df = process_data(df[(df['year'] > 2000)])
```
