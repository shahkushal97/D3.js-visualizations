```python
import pandas as pd
```

```python
df = pd.read_csv("./data.csv")
```

```python
df
```

```python
data_r = df[["recipient","commitment_amount_usd_constant","year"]]
```

```python
data_r.groupby(['recipient','year'], as_index=False).agg({"commitment_amount_usd_constant": "sum"})
```

```python
data_r = data_r.sort_values(by=['recipient', 'year'])
```

```python
data_r
```

```python
data_r = pd.pivot_table(data_r,values='commitment_amount_usd_constant',index='recipient',columns='year')
```

```python
data_r = data_r.fillna(0)
```

```python
#result = data.transpose() 
data_r
```

```python
data_r.to_csv("./q1_r.tsv",index=True,sep='\t')
```
